Ce notebook explore les facteurs influen√ßant les √©missions de CO2 des v√©hicules de la flotte en utilisant les donn√©es retrait√©es.

```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sys
import os

# Ajouter le dossier src au path
sys.path.append(os.path.abspath('../src'))
from preprocessing import load_co2_data

# Configuration style
plt.style.use('ggplot')
sns.set_palette("magma")
```

## 1. Chargement des donn√©es (Preprocessed)
La fonction `load_co2_data` g√®re l'encodage des types de carburant et des marques.

```python
DATA_PATH = '../data/CO2 Emissions_Canada.csv'
df, _ = load_co2_data(DATA_PATH)
df.head()
```

## 2. Taille du moteur vs √âmissions

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Engine Size(L)', y='CO2 Emissions(g/km)', hue='Cylinders', data=df, alpha=0.5)
plt.title('Relation entre Taille Moteur et √âmissions de CO2')
plt.show()
```

## 3. Impact de la Consommation Mixte

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Fuel Consumption Comb (L/100 km)', y='CO2 Emissions(g/km)', data=df, scatter_kws={'alpha':0.1})
plt.title('Consommation Combin√©e vs CO2')
plt.show()
```
